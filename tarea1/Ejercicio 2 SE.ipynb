```python
import pandas as pd
from scipy.stats import spearmanr
import statistics
import matplotlib.pyplot as plt
```

```python
# Importamos la base 
t = pd.read_excel("/Users/igabo/Desktop/Tarea 1 semi/Tarea 1/data_schools.xlsx")
t.head()
```

# n = 15

```python
w = []
n = 20
while n <= 300:
    z = []
    for j in range(n):
        
        # Muestra aleatoria de tamaño 15
        k = t["escuela"].sample(n=15, replace = True).to_list()
        
        # Valores de LSAT
        x = [t.loc[t["escuela"] == i]["LSAT"].to_list()[0] for i in k]
        
        # Valores de GPA
        y = [t.loc[t["escuela"] == i]["GPA"].to_list()[0] for i in k]
        
        # Correlación entre LSAT y GPA
        z.append(spearmanr(x,y)[0])
    
    # Error estándar
    v = []
    for j in range(len(z)):
        v.append((z[j]-statistics.mean(z))**2)
        
    l=(sum(v)/(len(z)-1))**(0.5)
        
    w.append(l)
    n += 1
w
```

```python
plt.hist(w, 20, color = "steelblue", ec = "black")
```
